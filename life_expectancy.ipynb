```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average life expectancy by country
avg_life_exp_by_country = df_cleaned.groupby('Country')['Life expectancy'].mean().reset_index()

# Sort by life expectancy descending
avg_life_exp_by_country = avg_life_exp_by_country.sort_values(by='Life expectancy', ascending=False)

# Select top 20 countries for clearer visualization
top_countries = avg_life_exp_by_country.head(20)

# Plot
plt.figure(figsize=(14, 10))
sns.barplot(data=top_countries, y='Country', x='Life expectancy', palette='viridis')

plt.title("Top 20 Countries by Average Life Expectancy", fontsize=18, weight='bold')
plt.xlabel("Average Life Expectancy", fontsize=14)
plt.ylabel("Country", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add value labels to each bar
for index, value in enumerate(top_countries['Life expectancy']):
    plt.text(value + 0.3, index, f"{value:.1f}", va='center', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate your dataset by country (e.g., average all numeric columns per country).
# Then calculate the correlation on this aggregated data.
# Finally, plot the heatmap on the aggregated correlation matrix.

# Aggregate numeric variables by country (average)
df_country_avg = df_cleaned.groupby('Country').mean(numeric_only=True).reset_index()

# Drop the 'Country' column before calculating the correlation matrix
corr_matrix = df_country_avg.drop(columns=['Country']).corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, fmt=".2f", square=True)

plt.title("Correlation Heatmap of Numeric Variables (Averaged Across Countries)", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate data by country: average GDP and Life Expectancy, keep Status
df_country_avg = df_cleaned.groupby('Country').agg({
    'GDP': 'mean',
    'Life expectancy': 'mean',
    'Status': 'first'  # Assuming status is consistent per country
}).reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_country_avg, x='GDP', y='Life expectancy', hue='Status', alpha=0.8, s=100)

plt.title("Average Life Expectancy vs GDP Across Countries", fontsize=16)
plt.xlabel("Average GDP", fontsize=12)
plt.ylabel("Average Life Expectancy", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
