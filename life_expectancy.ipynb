public health initiatives, and sustainable development goals. 

This project leverages a comprehensive dataset encompassing various health, economic, social determinants and demographic indicators to 
conduct a detailed analysis of life expectancy trends and drivers across the globe.


### Project Goal

The primary goal of this project is to thoroughly investigate and compare the dynamics of life expectancy in both developed
and developing countries, identifying the key factors that contribute to its variation and exploring the reciprocal relationships

Objective 4: Determining life expectancy across countries

This study aims to explore how various social,economic and health factors influence life expectancy across countries.

Correlation Heatmap

The purpose of this is to visually explore the relationships between different numeric variables in the Life Expectancy dataset by using a correlation heatmap, averaged across countries.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate numeric variables by country (average)
df_country_avg = df.groupby('Country').mean(numeric_only=True).reset_index()

# Calculate correlation matrix on aggregated data (exclude 'Country' column)
corr_matrix = df_country_avg.drop(columns=['Country']).corr()

plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, fmt=".2f", square=True)

plt.title("Correlation Heatmap of Numeric Variables (Averaged Across Countries)", fontsize=16)
plt.tight_layout()
plt.show()


```

HISTPLOT WITH ECONOMIC FACTORS

The histplot shows how life expectancy is distributed across countries, separated by their development status ie Developed and Developing

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df_country_avg, x='Life expectancy', bins=30, hue='Status', multiple='stack', alpha=0.7)

plt.title("Distribution of Average Life Expectancy by Status", fontsize=16)
plt.xlabel("Average Life Expectancy", fontsize=12)
plt.ylabel("Number of Countries", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

SCATTERPLOT WITH SOCIAL FACTORS

Education is widely recognized as a key social determinant of health. In this analysis, we explore the relationship between average years of schooling and life expectancy across different countries. By aggregating data per country and plotting a scatter plot differentiated by development status, we aim to uncover whether better education outcomes correlate with longer lives  and whether this trend differs between developed and developing nations.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate data by country: average Schooling and Life Expectancy, keep Status
df_country_avg = df.groupby('Country').agg({
    'Schooling': 'mean',
    'Life expectancy': 'mean',
    'Status': 'first'  # Assuming each country has one consistent status
}).reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_country_avg, x='Schooling', y='Life expectancy', hue='Status', alpha=0.8, s=100)

plt.title("Average Life Expectancy vs Schooling Across Countries", fontsize=16)
plt.xlabel("Average Schooling (years)", fontsize=12)
plt.ylabel("Average Life Expectancy", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

SCATTERPLOT WTH HEALTH FACTORS

HIV/AIDS is a major public health issue, particularly in low- and middle-income countries. In this visualization, we examine how the prevalence of HIV/AIDS in a country correlates with its average life expectancy. By plotting these two variables and distinguishing countries based on their development status, we aim to highlight the impact of infectious diseases on national health outcomes and longevity.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate data by country: average HIV/AIDS and Life Expectancy, keep Status
df_country_avg = df.groupby('Country').agg({
    'HIV/AIDS': 'mean',
    'Life expectancy': 'mean',
    'Status': 'first'  # Assuming Status is consistent for each country
}).reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_country_avg, x='HIV/AIDS', y='Life expectancy', hue='Status', alpha=0.8, s=100)

plt.title(" Average Life Expectancy vs HIV/AIDS Prevalence Across Countries", fontsize=16)
plt.xlabel("Average HIV/AIDS Prevalence", fontsize=12)
plt.ylabel("Average Life Expectancy", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Objective 5: Assessing the Impact of Health Interventions on Life Expectancy
This section aims to determine the effect of specific health interventions, such as vaccination coverage (e.g., Hepatitis B, Polio, Diphtheria) and overall healthcare expenditure, on improvements in life expectancy.

#### Step 1: Select Relevant Health-Related Columns
We isolate the Life expectancy column along with the health intervention variables for our analysis.
These variables include vaccination coverages and healthcare expenditures.

```python
# Select columns related to life expectancy and health interventions
health_df = df[['Life expectancy', 'Hepatitis B', 'Polio', 'Diphtheria',
                'Total expenditure', 'percentage expenditure', 'Status']]

# Display info for the selected DataFrame
print(health_df.info())
```

#### Step 2: Check Correlation with Life Expectancy
To understand the linear relationships between Life expectancy and the health intervention variables, we compute the Pearson correlation coefficients.

```python

# Compute correlation matrix
correlation_matrix = health_df.corr(numeric_only=True)

# Show correlations with life expectancy, sorted in descending order
correlation_with_life_exp = correlation_matrix['Life expectancy'].sort_values(ascending=False)
print("Correlation of Health Interventions with Life Expectancy:")
print(correlation_with_life_exp.drop('Life expectancy').to_markdown(numalign="left", stralign="left"))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
correlation_matrix = health_df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlation Heatmap: Life Expectancy and Health Interventions")
plt.tight_layout()
plt.show()
```

### Interpretation of the Correlation
The correlation analysis reveals the following relationships:

percentage expenditure (0.41): This variable shows the strongest positive correlation with Life expectancy. This suggests that countries that allocate a higher percentage of their total national expenditure to health tend to experience higher life expectancies.

Diphtheria (0.34) and Polio (0.33): Vaccination coverage for these diseases exhibits a moderate positive correlation with Life expectancy. This indicates that higher immunization rates are associated with longer life spans.

Hepatitis B (0.20) and Total expenditure (0.17): These variables show weaker positive correlations. While a positive relationship exists, their linear impact on Life expectancy appears less pronounced compared to percentage expenditure, Diphtheria, and Polio.

Overall, all listed health interventions demonstrate a positive association with Life expectancy, suggesting that investments in public health and vaccination programs are beneficial for human longevity.

### Scatter Plots:
We'll visualize the relationship between Life expectancy and the two expenditure variables, coloring points by country Status (Developed vs. Developing).



### Scatter plot for 'Life expectancy' vs. 'percentage expenditure', colored by 'Status'

```python
import altair as alt

# Create a scatter plot for 'Life expectancy' vs. 'percentage expenditure', colored by 'Status'
chart1 = alt.Chart(df).mark_point(filled=True, opacity=0.6).encode(
    x=alt.X('percentage expenditure', title='Percentage of Total Expenditure on Health'),
    y=alt.Y('Life expectancy', title='Life Expectancy (Years)'),
    color=alt.Color('Status', title='Country Status'),
    tooltip=['Country', 'Year', 'Status', 'Life expectancy', 'percentage expenditure']
).properties(
    title='Life Expectancy vs. Percentage of Total Expenditure on Health'
).interactive() # Make the chart interactive for zooming and panning

chart1.display()




```

### Interpretation of Life Expectancy vs. Percentage of Total Expenditure on Health: 
The plot generally shows a positive relationship. Developed countries tend to cluster in the upper-right portion, indicating both higher percentage expenditure and higher Life expectancy. Developing countries are more dispersed, with many occupying the lower expenditure and life expectancy ranges, though some show progress. This visually supports the stronger correlation observed for percentage expenditure

### Scatter plot for 'Life expectancy' vs. 'Total expenditure', colored by 'Status'

```python
# Create a scatter plot for 'Life expectancy' vs. 'Total expenditure', colored by 'Status'
chart2 = alt.Chart(df).mark_point(filled=True, opacity=0.6).encode(
    x=alt.X('Total expenditure', title='Total Expenditure on Health'),
    y=alt.Y('Life expectancy', title='Life Expectancy (Years)'),
    color=alt.Color('Status', title='Country Status'),
    tooltip=['Country', 'Year', 'Status', 'Life expectancy', 'Total expenditure']
).properties(
    title='Life Expectancy vs. Total Expenditure on Health'
).interactive()
chart2.display()
```

### Interpretation of 'Life Expectancy' vs 'Total Expenditure' on Health
A positive trend is also evident here, but the data points are more scattered compared to the percentage expenditure plot, consistent with its weaker correlation. Developed countries still tend to have higher values for both metrics, while Developing countries show a wider spread across lower Total expenditure values.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select relevant health-related columns for the heatmap
health_cols_for_heatmap = ['Life expectancy', 'Hepatitis B', 'Polio', 'Diphtheria',
                           'Total expenditure', 'percentage expenditure']
health_df_for_heatmap = df[health_cols_for_heatmap]

# Compute the correlation matrix
correlation_matrix_full = health_df_for_heatmap.corr()

# Create a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix_full, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Health Interventions and Life Expectancy')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.savefig('correlation_heatmap_health_interventions.png')
# plt.show() # In a real environment, you'd save without show() for report generation.
```

You can clearly see the positive relationships between Life expectancy and the health intervention variables.

It also shows correlations among the intervention variables themselves (e.g., how Hepatitis B relates to Polio, etc.), providing a holistic view of the interdependencies within the health data.
